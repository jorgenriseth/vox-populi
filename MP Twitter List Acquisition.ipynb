```python
import json
import requests
from bs4 import BeautifulSoup
import os

url = "https://www.mpsontwitter.co.uk/list"
data = requests.get(url)

html = BeautifulSoup(data.text, 'html.parser')
table = html.select("tbody", id='mp_wrapper')[1]

mp_dict = {}
for line in table.select('tr'):
    name = line.select('td')[2].get_text().strip()
    party = line.td["class"][0]
    twitter_id = line.a.get_text()[1:]
    url = line.a['href']
    mp_dict[name] = {
        "party": party,
        "twitter_id": twitter_id,
        "url": "https://twitter.com/" + twitter_id
    }
    
with open('mps.json', 'w') as f:
    json.dump(mp_dict, f)
```

```python
def get_tweet_timeline(url):
    """
    Takes in a url for a specific twitter user timeline, and stores all
    tweets in a dictionary. Dictionary key is the tweet id, while the values 
    are strings with the tweet text content.
    """
    data = requests.get(url)
    html = BeautifulSoup(data.text, 'html.parser')
    
    timeline = html.select('#timeline li.stream-item')
    all_tweets = {}
    
    # Loop through every tweet and store content in all_tweets
    # dictionary.
    
    for tweet in timeline:
        try:
            # Parse tweet with id
            tweet_id = tweet['data-item-id']
            tweet_text = tweet.select('p.tweet-text')[0].get_text()

            all_tweets[tweet_id] = tweet_text
        
        # Retweets span multiple tweets in current parsing, hence 
        # they should just be dropped. 
        except IndexError:
            pass
        
    return all_tweets
```

```python
user_tweets = {}
for mp in mp_dict:
    print('='*80)
    print(mp)
    user_tweets[mp] = get_tweet_timeline(mp_dict[mp]["url"])
```

```python
dir_tweets = 'tweets/'

for mp, tweets in user_tweets.items():
    # Create path-friendly filename
    filename = mp.replace(' ', '').lower() + '-'
    
    # Create a file containing all tweets for current mp.
    f_all = open(dir_tweets + 'all_tweets/' + filename + 'all.txt', 'w')
    
    # Create seperate files for each of the different tweets
    for tweet_id, text in tweets.items():
        f_all.write('\n')
        f_all.write(text)
        
        with open(dir_tweets + filename + tweet_id + '.txt', 'w') as f:
            f.write(text)
        
    f_all.close()
```
