```python
import requests
from bs4 import BeautifulSoup

url = "https://twitter.com/albertocostamp"
data = requests.get(url)
html = BeautifulSoup(data.text, 'html.parser')

timeline = html.select('#timeline li.stream-item')
all_tweets = []

timeline = timeline
i = 0
for tweet in timeline:
    try:
        # Parse tweet with id
        tweet_id = tweet['data-item-id']
        tweet_text = tweet.select('p.tweet-text')[0].get_text()

        # Create a list of hyperlink-tags
        hyperlinks = tweet.select('p.tweet-text')[0].select('a')

        # Check if hyperlink refers to image/video, and if so, strip from tweet text
        for link in hyperlinks:
            if 'pic.twitter' in link.get_text():
                tweet_text = tweet_text.strip(link.get_text())

        all_tweets.append({"id": tweet_id, "text": tweet_text})
    except IndexError:
        pass

for tweet in all_tweets:
    print('=' * 80)
    print(f"Id: {tweet['id']}")
    print(f"Text:\n{tweet['text']}")
          
```
